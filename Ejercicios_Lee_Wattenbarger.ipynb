### G.- Calcule el factor de compresibilidad - (z), densidad $ \rho_g $, compresibilidad $ c_g $ y la viscosidad $ \mu_g $ para un gas a 249 °F y presiones de 500, 1000, 1500, 2000, 2500 y 3000 psia. Muestre los resultados en una tabla o un dataframe.

```julia
Ty_e12 = 249
p_e12 = [i for i in 500:250:3500] ;
```

#### Inicializando los contenedores para los resultados

```julia
# z_e12 = empty([], Union{Int64, Float64})
# dg_e12 = empty([], Union{Int64, Float64})
# cg_e12 = empty([], Union{Int64, Float64})
# ug_e12 = empty([], Union{Int64, Float64})
# psr_e12 = empty([], Union{Int64, Float64})
```

#### Calcular la presión y temperatura seudoreducidas

```julia
psr_e12 = [p_e12[i]/Psc_corr_e12 for i in eachindex(p_e12)]
```

```julia
Tsr_e12 = (Ty_e12 + 460)/Tsc_corr_e12
```

### Factor de compresibilidad del gas - z

```julia
z_Papay_e12 = [z_Papay(psr_e12[i], Tsr_e12) for i in eachindex(p_e12)]
```

```julia
z_Sarem_e12 = [z_Sarem(psr_e12[i], Tsr_e12) for i in eachindex(p_e12)]
```

### Densidad del gas $ \rho_g $

```julia
dg_e12 = [round(2.7*p_e12[i]*Ge_gas_mix_e12/(z_Papay_e12[i]*Ty_e12), digits=4) for i in eachindex(p_e12)]
```

### Compresibilidad del gas $ c_g $

```julia
cg_Papay_e12 = [cg_Papay(psr_e12[i], Tsr_e12, z_Papay_e12[i], Psc_corr_e12) for i in eachindex(p_e12)]
```

```julia
cg_Sarem_e12 = [cg_Sarem(psr_e12[i], Tsr_e12, z_Sarem_e12[i], Psc_corr_e12) for i in eachindex(p_e12)]
```

### Viscosidad del gas - $ \mu _g $

```julia
ug_Carr_Kob_e12 = [ug_Carr_Kobayashi_Burrows(Ty_e12, psr_e12[i], Tsr_e12, M_nhc_e12, fm_hc_e12, fm_n2_e12, fm_co2_e12, fm_h2s_e12) for i in eachindex(p_e12)]
```

```julia
ug_Carr_Kob_e12 = [ug_Carr_Kobayashi_Burrows(Ty_e12, psr_e12[i], Tsr_e12, M_nhc_e12, fm_hc_e12, [fm_n2_e12, fm_co2_e12, fm_h2s_e12]) for i in eachindex(p_e12)]
```
